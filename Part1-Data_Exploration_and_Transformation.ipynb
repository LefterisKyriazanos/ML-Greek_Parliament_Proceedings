```python
x = []
y = []

for i in speeches['political_party'].unique():
    x.append(i)
    y.append(speeches[speeches['political_party']== i]['member_name'].nunique())
 

speakers = {'Politica Party':x,'Different Speakers':y}
speakers = pd.DataFrame(speakers)
speakers.sort_values(by='Different Speakers',ascending=False,inplace=True)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 20))
ax = sns.barplot(y="Politica Party", x="Different Speakers", 
                
                 edgecolor='black',
                 data=speakers,
                 orient='h')
```

```python
speeches['sitting_date'] =  pd.to_datetime(speeches['sitting_date'])
speeches['sitting_date'] = pd.DatetimeIndex(speeches['sitting_date']).year
speeches['sitting_date']
```

```python
speeches[speeches['political_party']== 'νεα δημοκρατια']['sitting_date'].value_counts(dropna=True)
```

```python

speeches_over_years = pd.DataFrame(columns=['political_party', 'Year', 'Speeches'])
for i in speeches['political_party'].unique():
    value_counts = speeches[speeches['political_party']== i]['sitting_date'].value_counts(dropna=True)
   


    df_val_counts = pd.DataFrame(value_counts)
    df_value_counts_reset = df_val_counts.reset_index()
    df_value_counts_reset.columns = ['Year', 'Speeches']
    df_value_counts_reset['political_party'] = i
    df_value_counts_reset.sort_values(by='Year',inplace=True)
    speeches_over_years = pd.concat([speeches_over_years, df_value_counts_reset], sort=False)

        
    
```

```python
bins = [0,1,5,10, 25, 100,400,1000,4000,42342]
speeches['binned'] = pd.cut(speeches['speech_length'], bins)

speeches['binned'].value_counts()
```

```python
df = speeches['binned'].value_counts().rename_axis('word count of speech').reset_index(name='speeches')
df
```

```python
speeches.loc[speeches['speech_length'] >=18,'speech'].count()
```

```python
data= [[]]
for i in range(1,4001):
    results = set()
    speeches.loc[speeches['speech_length'] ==i,'speech'].str.lower().str.split().apply(results.update)
    data +=  [[i,len(results)]]
unique_words = pd.DataFrame(data,columns=['speech_length','unique_words'])    
unique_words
```

number of unique words used for each lenght of speech.

```python
unique_words.sort_values(by='unique_words',ascending=False,inplace=True)
unique_words.dropna(axis=0,inplace=True)
unique_words = unique_words[unique_words.unique_words > 0]
unique_words.reset_index(drop=True,inplace=True)
unique_words
```
